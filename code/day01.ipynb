```
#tf.cond 예제
x=tf.constant(1.)
bool = tf.constant(True)
res = tf.cond(bool,lambda:tf.add(x, 1.),lambda: tf.add(x, 10.))

print(res)
#print(res.numpy())
```

```
x=tf.constant(2)
y=tf.constant(5)
def f1(): return tf.multiply(x,17)
def f2(): return tf.add(y,23)
r = tf.cond(tf.less(x,y), f1,f2)

print(r)
#r.numpy() 값은 34로 나와야한다
```

```
import numpy as np

print(np.arrange(3))
print(np.ones((3,3)))
print()

x=tf.constant((np.arrange(3)))
y=tf.constant([5], dtype=tf.int64)
print(x)
print(y)
print(x+y)
```

```
#브로드캐스팅 코드 2
x=tf.constant((np.arrange(3)))
y = tf.constant([5], dtype=tf.int64)
print((x+y).numpy())

x=tf.constant((np.ones((3,3))))
y=tf.constant(np.arrange(3), dtype=tf.double)
print((x+y).numpy())

x=tf.constant(np.arrange(3).reshape(3,1))
y = tf.constant(np.arrange(3))
print((x+y).numpy())
```
