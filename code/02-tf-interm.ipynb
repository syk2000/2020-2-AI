행렬 곱셈:딥러닝 배우면 꼭배워야한다.
ex) A의 뒤의 행과 B의 앞의 행의 값이 같아야한다
A : 2*3
b:  3*2

결과는 2*2가 되어야 한다. 행렬결과이기도 하고 행렬의 곱셈이기도 하다.

```
x = [[2,]]
m = tf.matmul(x,x)
print(m)
print(m.numpy())
```

```
#Matrix multiplications 1
matrix1 = tf.constant([[1.,2.],[3.,4.]])
matrix2 = tf.constant([[2.,0.],[1.,2.]])

gop = tf.matmul(matrix1, matrix2)
print(gop.numpy())
```

```
#Matrix multiplicaions 2

gop = tf.matmul(matrix2, matrix1)
print(gop.numpy())

#matrix1과 matrix2의 곱의 위치를 바꿔 하면 값이 다르다 즉 교환법칙이 성립하지 않는것이다.
```

Shape & reShape ****

```
rank_three_tensor = tf.ones([3, 4, 5])
rank_three_tensor.shape
```

```
rank_three_tensor.numpy()
```

```
#기존 내용을 6x10 행렬로 형태 변경
matrix = tf.reshape(rank_three_tensor, [6, 10])
matrix
```

```
#기존 내용을 3x20 행렬로 형태 변경
# -1은 차원 크기를 계산하여 자동으로 결정하라는 의미
matrixB = tf.reshape(matrix, [3,-1])
matrixB
```

```
#기존 내용을 4x3x5 텐서로 형태 변경
matrixAlt = tf.reshape(matrixB, [4, 3, -1])
matrixAlt
```

```
#형태가 변경된 텐서의 원소 개수는 원래 텐서의 원소 개수와 같다
# 그러므로 다음은 원소의 개수를 유지하며
#마지막 차원에서 사용가능한 수가 없기 때문에 에러를 발생
yet_another = tf.reshape(matrixAlt, [13,2,-1]) #에러!
```

```
a=2
b=3
c= tf.add(a,b)
print(c.numpy())
```

```
x=2
y=3
add_op = tf.add(x,y)
mul_op = tf.multiply(x,y)
pow_op = tf.pow(add_op, mul_op)

print(pow_op.numpy())
```

```
a=tf.constant(2.)
b=tf.constant(3.)
c=tf.constant(5.)

#Some more operations
mean = tf.reduce_mean([a,b,c])
sum = tf.reduce_sum([a,b,c])

print("mean = ",mean.numpy())
print("sum = ", sum.numpy())
```
