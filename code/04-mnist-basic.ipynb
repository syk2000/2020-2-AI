```
#MNIST 데이터 셋을 로드하여 준비
# 전처리
#샘플 값을 정수에서 부동소수로 변환 == 0~255== 부동소수
#한비트의 값을 255로 나눔 == 0~1

import tensorflow as tf

mnist = tf.keras.datasets.mnist

(x_train,y_train),(x_test, y_test) = mnist.load_data()

#샘플값을 정수(0~255)에서 부동소수(0~1)로 변환
x_train, x_test = x_train / 255.0, x_test / 255.0 # == 간단한 정규화로 알 수가 있다.
```

```
#층을 차례대로 쌓아 tf.keras.models.Sequential 모델을 생성
model = tf.keras.models.Sequential([
                                    tf.keras.layers.Flatten(input_shape=(28,28)),
                                    tf.keras.layers.Dense(128, activation='relu'),#128 = 뉴런 activation = 활성화 함수
                                    tf.keras.layers.Dropout(0.2),
                                    tf.keras.layers.Dense(10, activation='softmax')
])
#Dense와 Dense 사이를 연결시켜 주는 부분에 가중치와 편향이라는 것이 들어있다. == 실수이다 & 138개의 뉴런이 있는 것이다.
#Dense() 완전연결층 == 밀집층
```

옵티마이저
입력된 데이터와 손실 함수를 기반으로 모델(w,b)를 업데이트하는 메커니즘

손실함수
훈련 데이터에서 신경망의 성능을 측정하는 방법
모델이 옳은 방향으로 학습될 수 있도록 도와주는 기준 값

```
#층을 차례대로 쌓아 tf.keras.models.Sequential 모델을 생성

model = tf.keras.models.Sequential([
                                    tf.keras.layers.Flatten(input_shape=(28,28)),
                                    tf.keras.layers.Dense(128, activation='relu'),#128 = 뉴런 activation = 활성화 함수
                                    tf.keras.layers.Dropout(0.2),
                                    tf.keras.layers.Dense(10, activation='softmax')
])
```

```
#훈련에 사용할 옵티마이저 (optimizer)와 손실 함수,출력정보를 모델에 설정

model.compile(optimizer='adam',
              loss = 'sparse categorica_crossentropy',
              metrics=['accuracy'])
              #metrics=['accuracy', 'mse'])
#모델 요약 표시
model.summary()
#param = 가중치+편향수
```

```
#생성된 모델로 훈련 데이터 학습
#모델을 훈련 데이터로 총 5번 훈련
model.fit (x_train, y_train, epochs=5) # epochs = 손실값 = 작아지면 좋아짐
```

```
#테스트 데이터로 성능 평가
#모델을 테스트 데이터로 평가
model.evaluate(x_test, y_test)
```
