```
#5 와인 데이터셋 불러오기
import pandas as pd
red= pd.read_csv('http://archive.ics.uci.edu/ml/matchine-learning-databases/wine-quality/winequality-red.csv', sep=';')
white = pd.read_csv('http://archive.ics.uci.edu/ml/matchine-learning-databases/wine-quality/winequality-white.csv', sep=';')
print(red.head())
print(white.head())
```

```
red['type']=0
white['type'] = 1
print(red.head(2))
print(white.head(2))

wine = pd.concat([red, white])
print(wine.describe())
```

```
import matplotlib.pyplot as plt
plt.hist(wine['type'])
plt.xticks([0,1])
plt.show()

print(wine['type'].value_counts())
```
